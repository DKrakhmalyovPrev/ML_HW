Наивный байесовский классификатор

```python
from sklearn import datasets, cross_validation, naive_bayes
import numpy as np
from pandas import DataFrame
```

Загрузим датасеты с обучающей выборкой, выведем некоторые данные для наглядности

```python
digits = datasets.load_digits()
```

```python
digits
```

```python
digits_frame = DataFrame(digits.data)
digits_frame['target'] = digits.target
digits_frame.head()
```

```python
breast_cancer = datasets.load_breast_cancer()
```

```python
breast_cancer
```

```python
print (breast_cancer.feature_names)
```

```python
breast_cancer_frame = DataFrame(breast_cancer.data)
breast_cancer_frame.columns = breast_cancer.feature_names
breast_cancer_frame['target'] = breast_cancer.target
breast_cancer_frame.head()
```

```python

breast_cancer_frame.target = breast_cancer_frame.target.apply(lambda x: breast_cancer.target_names[x])
breast_cancer_frame.head()
```

```python
bernoulli = naive_bayes.BernoulliNB()
multinom = naive_bayes.MultinomialNB()
gaus = naive_bayes.GaussianNB()
```

```python
accuracies_digits = {'BernoulliNB': 0, 'MultinomialNB' : 0, 'GaussianNB': 0}

accuracies_digits['BernoulliNB'] = cross_validation.cross_val_score(bernoulli, digits.data, digits.target).mean()
accuracies_digits['MultinomialNB'] = cross_validation.cross_val_score(multinom, digits.data, digits.target).mean()
accuracies_digits['GaussianNB'] = cross_validation.cross_val_score(gaus, digits.data, digits.target).mean()
```

```python
accuracies_digits
```

```python
accuracies_cancer = {'BernoulliNB': 0, 'MultinomialNB' : 0, 'GaussianNB': 0}

accuracies_cancer['BernoulliNB'] = cross_validation.cross_val_score(bernoulli, 
                                                                    breast_cancer.data, 
                                                                    breast_cancer.target).mean()
accuracies_cancer['MultinomialNB'] = cross_validation.cross_val_score(multinom, 
                                                                      breast_cancer.data, 
                                                                      breast_cancer.target).mean()
accuracies_cancer['GaussianNB'] = cross_validation.cross_val_score(gaus, 
                                                                   breast_cancer.data, 
                                                                   breast_cancer.target).mean()
```
